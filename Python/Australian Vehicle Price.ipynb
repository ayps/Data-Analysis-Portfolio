**using ANOVA to analyse if car brand has influence on price**

```python
from scipy.stats import f_oneway
```

```python
brand_price_analysis=f_oneway(*[df_cleaned['Price'][df_cleaned['Brand']==brand] for brand in df_cleaned['Brand'].unique()])
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
model = ols('Price ~ Brand', data=df_cleaned).fit()
anova_result = anova_lm(model)
print("ANOVA Table:")
print(anova_result)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
tukey_result = pairwise_tukeyhsd(endog=df_cleaned['Price'], groups=df_cleaned['Brand'], alpha=0.05)
```

```python
tukey_df = pd.DataFrame(data=tukey_result._results_table.data[1:], columns=tukey_result._results_table.data[0])
```

```python
tukey_df
```

```python
tukey_df[tukey_df['reject']==True]
```

```python
model = ols('Price ~ DriveType', data=df_cleaned).fit()
anova_result_DriveType = anova_lm(model)
print("ANOVA Table:")
print(anova_result_DriveType)
```

```python
from scipy.stats import ttest_ind
```

```python
automatic_prices = df_cleaned[df_cleaned['Transmission'] == 'Automatic']['Price']
manual_prices = df_cleaned[df_cleaned['Transmission'] == 'Manual']['Price']
t_statistic, p_value = ttest_ind(automatic_prices, manual_prices, equal_var=False)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
